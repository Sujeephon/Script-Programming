```python
print("Choose your path: LEFT or RIGHT")

choice1 = input("Enter your choice: ").lower()

if choice1 == "left":
    print("\nYou found a mysterious treasure room!")
    print("Do you want to open it?")

    choice2 = input("Yes/No: ").lower()

    if choice2 == "yes":
        print("\nCongratulations!")
        print("You found the treasure and safely!")
    elif choice2 == "no":
        print("\nYou walked away safely but missed the treasure(extra point).")

elif choice1 == "right":
    print("\nA zombie suddenly appears!")
    print("Do you want to RUN or Fight?")

    choice2 = input("Enter your choice: ").lower()

    if choice2 == "run":
        print("\nThe zombie caught you.")
        print("Game Over.")
    elif choice2 == "fight":
        print("\nThe zombie couldn't fight you.")
        print("You Win!")
    else:
        print("\nInvalid choice.")
        print("Game Over.")

else:
    print("\nInvalid choice.")
    print("Please restart the game.")
```
