```python
temp = 25
is_raining = True

if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")
```

if

ตรวจสอบเงื่อนไขแรก : temp > 20 (25 > 20) = True
ตรวจสอบเงื่อนไขที่ 2 : not is_raining  เมื่อใช้ not ก็จะเป็น not True = False
เมื่อใช้ตัวดำเนินการ AND ก็จะได้ True and False ผลลัพธ์จึงเป็น False โปรแกรมจึงไม่เข้าเงื่อนไข if 

elif

ตรวจสอบเงื่อนไขแรก : temp <= 20 (25 <= 20) = False
ตรวจสอบเงื่อนไขที่ 2 : is_raining = True
เมื่อใช้ตัวดำเนินการ OR ก็จะได้ False or True ผลลัพธืจึงเป็น True โปรแกรมจึงเข้าเงื่อนไข elif 
แสดงลัพธ์ Stay indoors or bring an umbrella.


```python
phrase = "Python 2026 Engine"

consonants = {
    c.lower()
    for c in phrase
    if c.isalpha() and c.lower() not in 'aeiou'
}

print(f"Unique consonants: {consonants}")
```

ตรวจสอบว่า c เป็นตัวอักษรภาษาอังกฤษไหม และ แปลงตัวอักษรให้เป็นพิมพ์เล็ก ตรวจสอบว่าไม่ใช่สระ

แสดงผลลัพธ์ Unique consonants: {'h', 'n', 'g', 'y', 'p', 't'}
