```python
# Square Pattern

for i in range(5):
    for j in range(5):
        print("*", end="")
    print()
```

```python
# Right-Angled Triangle

for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
    
```

```python
# Inverted Right-Angled Triangle

for i in range(5, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```
